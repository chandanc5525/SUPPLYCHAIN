```python
# Import Data using URL from GITHUB
url = 'https://raw.githubusercontent.com/chandanc5525/SUPPLYCHAIN/refs/heads/main/SCM.csv'

df = pd.read_csv(url)

df.sample(frac = 1) # Data Shuffle
```
